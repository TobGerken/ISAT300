```python
# running this will import pandas and numpy.
import pandas as pd
import numpy as np
```

```python
DaleEnterPriseData = 'https://raw.githubusercontent.com/TobGerken/ISAT300/main/Data/DaleEnterprise_TMINTMAXTOBS_20221023.csv'

df= pd.read_csv(DaleEnterPriseData)
print(df.shape)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df['DateObject']=pd.to_datetime(df['DATE'])
df['Month'] = df['DateObject'].dt.month
df['Day']= df['DateObject'].dt.day
df['Year']= df['DateObject'].dt.year

```

```python
# Eliminate all feb 29 for consistency. 
df = df[(df['Month']==2) & (df['Day']==29)]
df.head()
```

```python
df=df[((df['Month']==2) & (df['Day']==29)) == False]
df.head()
```

```python
df_new = df.pivot(index='Year', columns=['Month', 'Day'], values='TMAX')
```
