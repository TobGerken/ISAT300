```python
import matplotlib.pyplot as plt
import numpy as np

plt.rc('xtick', labelsize=14) 
plt.rc('ytick', labelsize=14) 

n=25
y1 = np.random.normal(loc=10, scale=.5, size=n)
y2 = np.random.normal(loc=60, scale=.5, size=n)

x1 = np.empty(n)
x1.fill(0)

x2 = np.empty(n)
x2.fill(100)
plt.style.use('seaborn-whitegrid')

fig,ax =  plt.subplots(figsize = (6, 4))

plt.scatter(x1,y1, color = 'lightgrey', marker = '.')
plt.scatter(x2,y2, color = 'lightgrey', marker = '.')
plt.plot([x1,x2],[10, 60], color = 'tab:blue')

ax.set_yticks(np.arange(0,65,10)) 
ax.set_xticks(np.arange(0,110,10))
ax.set_ylim([-2.5,65]) 
ax.set_xlim([-2.5,105]) 
ax.set_ylabel('Thermocouple Temperature [$^o$C]', fontsize = 14)
ax.set_xlabel('Actual Temperature [$^o$C]', fontsize = 14)
```
