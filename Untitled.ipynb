```python
import matplotlib.pyplot as plt
import numpy as np
fig = plt.figure(figsize = (6, 4))
a = 1000
x = 0
plt.plot(x,a,color = 'w', marker = '.',linestyle = '')
plt.ylim((0,100))
plt.xlim((0,100))
plt.ylabel(r'Conductivity ($\mu$S/cm)',fontsize =16)
plt.xlabel(r'Time (s)',fontsize =16)
plt.xticks(np.arange(0,100,5), color = 'w')
plt.yticks(np.arange(0,100,5), color = 'w')
plt.grid(True)
```
