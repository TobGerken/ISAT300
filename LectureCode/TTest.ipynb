```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
from scipy.stats import t
plt.style.use('ggplot')

```

```python
x = np.arange(55, 80, 0.1)
#plot normal distribution with mean 0 and standard deviation 1
fig = plt.figure(figsize = (6, 5))
ax = fig.add_subplot(1, 1, 1)

plt.plot(x, norm.pdf(x, 67.3, 2.7))
plt.plot(x, norm.pdf(x, 70.1, 3.0))
plt.vlines(67.3,0,0.16,color = 'r', linestyle = ':')
plt.vlines(70.1,0,0.16,color = 'b', linestyle = ':')
plt.title('Comparison of Height Distributions')

ax.set_ylabel('Probability',fontsize = 12, fontweight='bold')
ax.set_xlabel('Height',fontsize = 12, fontweight='bold')
ax.legend(['1914 cohort', '1997 cohort'], fontsize = 12)
plt.savefig('StdHeight.jpg')
```

```python
x = np.arange(55, 80, 0.1)
#plot normal distribution with mean 0 and standard deviation 1
fig = plt.figure(figsize = (6, 5))
ax = fig.add_subplot(1, 1, 1)

plt.plot(67.3, 1, color = 'r')
plt.plot(70.1, 1.5, color = 'b')
 
x_error = 0.3
    
plt.errorbar(67.3, 1,
             xerr = x_error,
             fmt ='o')

plt.errorbar(70.1, 1.5,
             xerr = x_error,
             fmt ='o')


ax.set_ylim((0,4))
ax.set_xlim((60,75))
ax.set_yticks(np.array([])) 

plt.title('Comparison of Standard Erorr')

ax.set_xlabel('Height',fontsize = 12, fontweight='bold')
ax.legend(['1914 cohort', '1997 cohort'], fontsize = 12)
```

```python
placebo = np.random.normal(217.4,30.3,15)
newdrug = np.random.normal(195.9,28.7,15)

```

```python
fig = plt.figure(figsize = (6, 5))
ax = fig.add_subplot(1, 1, 1)
plt.hist(placebo, alpha = 0.5)
plt.hist(newdrug, alpha = 0.5)

plt.vlines(217.4,0,4,color = 'r', linestyle = ':')
plt.vlines(195.9,0,4,color = 'b', linestyle = ':')

plt.legend(['Placebo', 'New Drug'])

ax.set_ylabel('Samples',fontsize = 12, fontweight='bold')
ax.set_xlabel('Cholesterol',fontsize = 12, fontweight='bold')
```

```python
x = np.arange(-10, 10, 0.1)
#plot normal distribution with mean 0 and standard deviation 1
fig = plt.figure(figsize = (8, 6))
ax = fig.add_subplot(1, 1, 1)

for i in np.arange(1.5, 3, 0.1):
    plt.plot(x, norm.pdf(x, -3, i), color = 'tab:red')
    
plt.plot(x, norm.pdf(x, 3, 1.5), linestyle='-', linewidth = 3, color = 'tab:blue')
ax.set_ylim((0,0.3))


ax.set_ylabel('Probability',fontsize = 14, fontweight='bold')
ax.set_xlabel('Cholesterol',fontsize = 14, fontweight='bold')

ax.set_ylim((0,0.3))
plt.savefig('CD1D2_3.jpg')
```
