```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
from scipy.stats import t
plt.style.use('ggplot')

```

```python
x = np.arange(55, 80, 0.1)
#plot normal distribution with mean 0 and standard deviation 1
fig = plt.figure(figsize = (6, 5))
ax = fig.add_subplot(1, 1, 1)

plt.plot(x, norm.pdf(x, 67.3, 2.7))
plt.plot(x, norm.pdf(x, 70.1, 3.0))
plt.vlines(67.3,0,0.16,color = 'r', linestyle = ':')
plt.vlines(70.1,0,0.16,color = 'b', linestyle = ':')
plt.title('Comparison of Height Distributions')

ax.set_ylabel('Probability',fontsize = 12, fontweight='bold')
ax.set_xlabel('Height',fontsize = 12, fontweight='bold')
ax.legend(['1914 cohort', '1997 cohort'], fontsize = 12)
```

```python
x = np.arange(55, 80, 0.1)
#plot normal distribution with mean 0 and standard deviation 1
fig = plt.figure(figsize = (6, 5))
ax = fig.add_subplot(1, 1, 1)

plt.plot(67.3, 1, color = 'r')
plt.plot(70.1, 1.5, color = 'b')
 
x_error = 0.3
    
plt.errorbar(67.3, 1,
             xerr = x_error,
             fmt ='o')

plt.errorbar(70.1, 1.5,
             xerr = x_error,
             fmt ='o')


ax.set_ylim((0,4))
ax.set_xlim((60,75))
ax.set_yticks(np.array([])) 

plt.title('Comparison of Standard Erorr')

ax.set_xlabel('Height',fontsize = 12, fontweight='bold')
ax.legend(['1914 cohort', '1997 cohort'], fontsize = 12)
```
