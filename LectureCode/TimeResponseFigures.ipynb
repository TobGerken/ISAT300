```python
t = np.arange(0, 5, 0.025)+3

T = 3
tau = 1
K = 1
a = 25
offs = 70
b = (offs-a)*K

def fun(t,a, b, T, tau):
    y = a + b * np.exp(-(t-T)/tau)
    return y

plt.rc('xtick', labelsize=14) 
plt.rc('ytick', labelsize=14) 

y = fun(t, a, b, T, tau)

y_uc = y + np.random.normal(loc=0, scale=2, size=len(y))

fig = plt.figure(figsize = (7, 5))

ax1= plt.plot(t,y, linewidth = 3, color ='r')
ax1 = plt.scatter(t, y_uc, color = 'silver', marker = '.')
ax1= plt.plot([0, 3],[70, 70], linewidth = 3, color ='r')

plt.xlim((2.75,8))

plt.xlabel('Time (seconds)',fontsize =16)
plt.ylabel(r'Temperature ($^o$C)',fontsize =16)

plt.text(4,60, r'$y(t) = A + B e ^{(-[t-T]/\tau)}$',fontsize = 24)
fig.savefig('TResp.jpg')
```
